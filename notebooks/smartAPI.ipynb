```python
predicate_list = list(mkg_df["predicate"].unique())
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# create a directed graph  -- all available relations 
G=nx.from_pandas_edgelist(mkg_df, "subject", "object", edge_attr=True,create_using=nx.MultiDiGraph())
```

```python
plt.figure(figsize=(12,12))

pos = nx.spring_layout(G)
nx.draw(G, with_labels=True, node_color='red', edge_cmap=plt.cm.Blues, pos = pos)
plt.show()
```

```python
# create a directed graph  -- one relation
G=nx.from_pandas_edgelist(mkg_df[mkg_df['predicate']=='related_to'], "subject", "object", edge_attr=True,create_using=nx.MultiDiGraph())
plt.figure(figsize=(12,12))
pos = nx.spring_layout(G, k = 0.5) # k regulates the distance between nodes
nx.draw(G, with_labels=True, node_color='blue', node_size=1500, edge_cmap=plt.cm.Blues, pos = pos, font_weight='bold')
plt.show()
```

```python
def metakg_graph_render(TRAPI_ID = None, predicate=None):
    """
    Generate a graph based on target predicate and write to graphml
    Pass to Marco (frontend)
    Questions:
        1. What if no predicate passed, graph all? (default case)
        2. What if no ID found, run large rendering?
            a. can run a fetch all but it is LARGE --dismissing this, too long and not useful?
        3. What API endpoint should this react to? (/api/metakg/....?)
            - adding to endpoint /format=graphml
        4. The answer to 3 will help me decide where to place this next
        5. Standardize output name for graphml
            a. something like [ID]_[PREDICATE]_.graphml
    """
    # we are going to take fields from the URL and generate a graph from the predicate to pass to the frontend 
    from controller import SmartAPI
    import pandas as pd
    import networkx as nx
    import matplotlib.pyplot as plt

    if TRAPI_ID: # do we need to add a check here? (if no ID, can't provide, raise error?)
        bte = SmartAPI.get(TRAPI_ID)
        mkg = bte.get_metakg()
        mkg_df = pd.DataFrame(mkg)
    # else:
    #     # -if we want to run this / keep as an option, it will take a minutefetch_all = SmartAPI.fetch_all_metakg()
    #     # after testing this, I am excluding it for now, it takes too long and might never be useful?
    #     fetch_all = SmartAPI.fetch_all_metakg()
    #     mkg_df = pd.DataFrame(fetch_all)

    mkg_df.drop(["provided_by"], axis=1, inplace=True)
    mkg_df.drop(["api"], axis=1, inplace=True)

    if predicate:
        outfile_name = f"{TRAPI_ID}_{predicate}_graph"
        # generate graph
        G=nx.from_pandas_edgelist(mkg_df[mkg_df['predicate']==predicate], "subject", "object", edge_attr=True,create_using=nx.MultiDiGraph())
        # write graph to graphml to pass forward to frontend
        nx.write_graphml(G,f'{outfile_name}.graphml')
    else:
        # no predicate passed
        outfile_name = f'{TRAPI_ID}_graph'
        # generate graph
        G=nx.from_pandas_edgelist(mkg_df, "subject", "object", edge_attr=True,create_using=nx.MultiDiGraph())
        # write graph to graphml to pass forward to frontend
        nx.write_graphml(G,f'{outfile_name}.graphml')

    # plot for testing
    plt.figure(figsize=(12,12))
    pos = nx.spring_layout(G, k = 0.5) # k regulates the distance between nodes
    nx.draw(G, with_labels=True, node_color='purple', node_size=1500, edge_cmap=plt.cm.Blues, pos = pos, font_weight='bold')
    plt.show()
```

```python
test_id = "dc91716f44207d2e1287c727f281d339"
test_predicate = "related_to"
metakg_graph_render(test_id, test_predicate)
```
